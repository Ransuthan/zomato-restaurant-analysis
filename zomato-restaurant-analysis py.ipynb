```python
import pandas as pd
df = pd.read_csv(r"C:\Users\ranju\Documents\data sets\zomato.csv", encoding='latin-1')
df.head(3)
```

```python
print(df.shape)
print(df.columns.tolist())
```

```python
df.info()
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')
df['Aggregate rating'].replace(0, float('nan'), inplace=True)
```

```python
for col in ['Has Table booking','Has Online delivery','Is delivering now']:
    df[col] = df[col].map({'Yes': 1, 'No': 0})
```

```python
df.head(3)
```

```python
df['Country Code'].value_counts().head(10)
```

```python
df['City'].value_counts().head(15)
```

```python
cuisines = df['Cuisines'].str.split(', ').explode()
cuisines.value_counts().head(15)
```

```python
import matplotlib.pyplot as plt
df['Aggregate rating'].dropna().hist(bins=20, edgecolor='white')
plt.title('Rating distribution')
plt.show()
```

```python
df.groupby('Has Online delivery')['Aggregate rating'].mean()
```

```python
df.groupby('Price range')['Aggregate rating'].mean()
```

```python
top_cuisines = cuisines.value_counts().head(10)
top_cuisines.plot(kind='barh', color='steelblue')
plt.title('Top 10 cuisines')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
sns.boxplot(x='Price range', y='Aggregate rating', data=df)
plt.title('Rating by price range')
plt.show()
```

```python
df['Has Online delivery'].value_counts().plot(kind='pie',
labels=['No delivery','Has delivery'], autopct='%1.1f%%')
plt.title('Online delivery adoption')
plt.show()
```

```python
df.groupby('Country Code')['Average Cost for two'].mean().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title('Avg cost for two by country')
plt.show()
```

```python
print(f"Online delivery rate: {df['Has Online delivery'].mean()*100:.1f}%")
print(f"Table booking rate: {df['Has Table booking'].mean()*100:.1f}%")
print(f"Avg rating overall: {df['Aggregate rating'].mean():.2f}")
```
